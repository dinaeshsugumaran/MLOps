Target Variable Analysis:

What is the distribution of the target variable, SalePrice? Is it skewed?

Are there any outliers in the SalePrice distribution?

How does SalePrice relate to key features like OverallQual, GrLivArea, and TotalBsmtSF?

```python
# What is the distribution of the target variable, SalePrice? Is it skewed?

sns.histplot(housing_data.SalePrice, kde= True, color= 'Red')
```

```python
#Are there any outliers in the SalePrice distribution?

plt.boxplot(housing_data.SalePrice,showmeans=True,showbox=True, showfliers=True)
plt.show()
```

```python
housing_data.SalePrice.describe()
```

```python
# To filter the outlier records
Q1 = np.percentile(housing_data.SalePrice, 25)
Q3 = np.percentile(housing_data.SalePrice, 75)
IQR = Q3-Q1
Max = Q3+(1.25*IQR)
```

```python
Max
```

```python
housing_data[housing_data['SalePrice']> Max]
```

```python
len(housing_data[housing_data['SalePrice']> Max])
#there are 184 records which are higher price than the max value from boxplot
```

```python
# How does SalePrice relate to key features like OverallQual, GrLivArea, and TotalBsmtSF?

sns.pairplot(housing_data[['SalePrice','Overall_Qual','Gr_Liv_Area','Total_Bsmt_SF']])
```

```python
#Calculate correlation coefficients
correlations = housing_data[['SalePrice','Overall_Qual','Gr_Liv_Area','Total_Bsmt_SF']].corr()
print(correlations['SalePrice'])
```

```python
#we can find some insigts using groupby


Overall_Qual_stats = housing_data.groupby('Overall_Qual')['SalePrice'].agg(['mean','median']).sort_values('mean',ascending=False)
Overall_Qual_stats['mean'] = Overall_Qual_stats['mean'].apply(lambda x: f"{x:.2f}").astype('float64')
Overall_Qual_stats['diff'] = Overall_Qual_stats['mean'] - Overall_Qual_stats['median']
Overall_Qual_stats

```

```python
neibhoruhood_stats = housing_data.groupby('Neighborhood')['SalePrice'].agg(['mean','median']).sort_values('mean',ascending=False)
neibhoruhood_stats['mean'] = neibhoruhood_stats['mean'].apply(lambda x: f"{x:.2f}")
neibhoruhood_stats
```

Numerical Variable Analysis:

Which numerical features are highly correlated with SalePrice?

Are there any pairs of numerical variables that are highly correlated with each other? What does this imply?

What is the distribution of key numerical features such as LotArea, YearBuilt, and 1stFlrSF?

```python
#Which numerical features are highly correlated with SalePrice?

numeric_columns = housing_data.describe().columns
numeric_columns
```

```python
numeric_columns = numeric_columns[3:]
```

```python
len(numeric_columns)
```

```python
corr_numeric = housing_data[numeric_columns].corr()
corr_numeric['SalePrice'].sort_values(ascending=False)
```

```python
#Are there any pairs of numerical variables that are highly correlated with each other? What does this imply?
plt.figure(figsize=(10, 8))
sns.heatmap(housing_data[numeric_columns].corr(), cmap='coolwarm')
```

```python

#sns.heatmap(housing_data[numeric_columns].corr(), cmap='coolwarm')

correlation_values = housing_data[numeric_columns].corr()
```

```python
threshold = 0.8

correlation_values_max_threshold = correlation_values[(correlation_values > threshold) | (correlation_values < -threshold )]
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_values_max_threshold, cmap='coolwarm',annot=True)
```
