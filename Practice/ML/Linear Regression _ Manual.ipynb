```python
#Linear Regression using Classes and methods
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

class linearRegression_manual():
    
    def __init__(self, learning_rate, no_of_iterations):
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations

    def fit(self, x,y):
        self.r, self.c = x.shape
        self.m = np.zeros(self.c)
        self.b = 0
        self.x = x
        self.y = y
        for i in range(self.no_of_iterations):
            self.update_values()
    def update_values(self):
        self.y_prediction = self.predict()
        dw = - (2 * (self.x.T).dot(self.y - self.y_prediction)) /self.r

        db = - 2 * np.sum(self.y - self.y_prediction)/self.r
    
        self.m = self.m - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db
    def predict(self):
        return self.x.dot(self.m)+self.b
        
    
```

```python
data = pd.read_excel('salary.xlsx')
X = data.iloc[:,:-1].values
Y = data.iloc[:,1].values

```

```python
lr_1 = linearRegression_manual(0.001,10000)
```

```python
lr_1.fit(X,Y)
```
