```python
data = pd.read_excel('salary.xlsx')
x = data.iloc[:,:-1].values
y = data.iloc[:,1].values

r, c = x.shape

learning_rate = .001
no_of_iterations = 7000
m = np.zeros(c)
b = 0

for i in range(no_of_iterations):
    y_prediction = x.dot(m)+b
    
    dw = - (2 * (x.T).dot(y - y_prediction)) /r

    db = - 2 * np.sum(y - y_prediction)/r
    
    m = m - learning_rate * dw
    b = b - learning_rate * db
```
